# clean text

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

stopwords = set(stopwords.words('english'))
```

```python
def clean(text):
    #lower casing
    text = text.lower()
    
    
    #removing punctuation, specila chars etc
    text = re.sub(r"[^a-zA-Z0-9\s]","",text)
    #tokenization
    text = word_tokenize(text)
    
    #stopwords removing
    text = [w for w in text if w not in stopwords]
    
    return " ".join(text)
```

```python
text = "Machine! learning enables 58998 @#$$$ computers to ? learn from data without being explicityly programmed."
clean(text)
```

```python
df['cleaned_abstract'] = df['abstract'].apply(clean)
```

```python
#Explanation(bellow code)
#Nouns: Captures both common and proper Nouns (singuer and plural)
#adjectives : comparative(jjr) and superlative(jjs) base adjective(jj)
#verb: base, past, gerund,participle
#adverb :competitave(rbr) and superlative(rbs) standard adverbs(rb)
```

```python
from textblob import TextBlob
from collections import Counter

def get_keywords(text):
    #creae a TextBlob object
    blob = TextBlob(text)
    
    #Extract a broader range of keywords based on POS tagging
    keywords = [word for word, tag in blob.tags if tag in (
    'NN','NNS','NNP','NNPS',#nouns
        'JJ','JJR','JJS'#adjectives
        'RB','RBR','RBS'#adverbs
    )]
    
    
    
    #count the most common keywords
    keyword_counts = Counter(keywords)
    most_common_keywords = keyword_counts.most_common(5)
    return most_common_keywords
```

```python
df['keywords'] = df['cleaned_abstract'].apply(get_keywords)
```
